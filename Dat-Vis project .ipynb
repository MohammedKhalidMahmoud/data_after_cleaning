```python
#import libraries 
import pandas as pd
import numpy as np
```

```python
#read the data
ds=pd.read_csv('tmdb-movies.csv')
```

```python
ds.head()
```

```python
#our information about data
ds.info()
```

```python
#Detect the missing columns
ds.isnull()
```

```python
ds.isnull().sum()
```

```python
#remove missing columns
remove=['imdb_id','cast','homepage','director','tagline','keywords','overview','genres','production_companies']
dc=ds.drop(remove,axis=1)
```

```python
dc
```

```python
dr=dc.dropna(axis=0)
```

```python
dr.isna().sum()
```

```python
dr.duplicated().sum()
```

```python
dr.drop_duplicates(inplace=True)
dr.duplicated().sum()
```

```python
dr.dropna(axis=0,how='all')
```

```python
dr.isna().sum()
```

```python
dr
```

```python
dr.isnull()
```

```python
dr.describe()
```

```python
#replace 0 in revenue and budget with Nan
dr['budget']=dr['budget'].replace(0,np.NaN)
dr['revenue']=dr['revenue'].replace(0,np.NaN)
dr['budget_adj']=dr['budget_adj'].replace(0,np.NaN)
dr['revenue_adj']=dr['revenue_adj'].replace(0,np.NaN)



```

```python
dr
```

```python
# remove rows with nan values
dr= dr[dr['budget'].notna()]
dr= dr[dr['revenue'].notna()]
dr= dr[dr['revenue_adj'].notna()]
dr= dr[dr['budget_adj'].notna()]
```
